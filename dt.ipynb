```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
import random
```

```python
random.seed(100) # the purpose of setting a seed is to make the random number generation deterministic, meaning that if you run the same sequence of random functions after setting the same seed, you will get the same results every time
data = pd.read_csv('dt_data.csv')
print(data)
print(data.describe())
```

```python
x= data.iloc[:, [2,3]].values # select all rows, select the 2nd and 3rd columns
y= data.iloc[:, -1].values  # select the last column
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)  
# create regressor object
# Create Decision Tree classifier object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)

# Predict the response for test dataset
y_pred = clf.predict(x_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Decision tree computing 'gini'
from sklearn import tree
import matplotlib.pyplot as plt
X, y = x_train, y_train
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
# Plotting the decision tree
plt.figure(figsize=(12,12))
tree.plot_tree(clf, fontsize=10)
plt.show()
```
